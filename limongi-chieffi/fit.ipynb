```python

from numpy import genfromtxt,polyfit,poly1d,log10
import matplotlib.pyplot as plt


data = genfromtxt("table-3.dat",delimiter=",")

mass = data[:,0]
al_wind = data[:,2]
al_sne  = data[:,4]
fe_sne  = data[:,10]

al_wind_fit = polyfit(mass,log10(al_wind),5)
al_wind_1d  = poly1d(al_wind_fit)

print(al_wind_fit)

al_sne_fit = polyfit(mass,log10(al_sne),3)
al_sne_1d  = poly1d(al_sne_fit)

print(al_sne_fit)


fe_sne_fit = polyfit(mass,log10(fe_sne),3)
fe_sne_1d  = poly1d(fe_sne_fit)

print(fe_sne_fit)


plt.semilogy(mass,al_wind,label="Al26 Wind")
plt.semilogy(mass,10**al_wind_1d(mass),label="Al26 Wind Fit")
plt.semilogy(mass,al_sne,label="Al26 SNe")
plt.semilogy(mass,10**al_sne_1d(mass),label="Al26 SNe Fit")
plt.semilogy(mass,fe_sne,label="Fe60 SNe")
plt.semilogy(mass,10**fe_sne_1d(mass),label="Fe60 SNe Fit")

plt.legend()

```
